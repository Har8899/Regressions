```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the Dataset

data = pd.read_csv('Position_Salaries.csv')
data.head()
```

```python
# Splitting the data into Independent and dependent variable
# Independent Variable 
X = data.iloc[:,1:-1].values
#Dependent Variable
y = data.iloc[:,-1].values
```

```python
# training the Linear Regression model on the whole data set

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

```python
# training the Polynomial Regression model on the whole dataset

from sklearn.preprocessing import PolynomialFeatures
pr = PolynomialFeatures(degree = 4)
X_poly = pr.fit_transform(X)
lr_2 = LinearRegression()
lr_2.fit(X_poly, y)
```

```python
# Visualizing the Linear Regression results

plt.scatter(X,y, color ='red')
plt.plot(X,lr.predict(X),color = 'blue')
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualizing Polynomial Regression 

plt.scatter(X, y, color ='r')
plt.plot(X, lr_2.predict(pr.fit_transform(X)),color = 'b')
plt.title('Truth or Bluff (Polynomial regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Polynomial Regression results(For higher resolution and smoother curve)
x_grid = np.arange(min(X),max(X),0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(X,y , color ='r')
plt.plot(x_grid,lr_2.predict(pr.fit_transform(x_grid)), color = 'b')
plt.title('Truth or Bluff (Polynomial regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# predicting a new result with Linear Regression

lr.predict([[6.5]])
```

```python
# predicting a new result with polynomial Regression

lr_2.predict(pr.fit_transform([[6.5]]))
```
